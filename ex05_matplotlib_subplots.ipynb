```python
import matplotlib.pyplot as plt
```

```python
# Create Subplots: 2 rows, 3 columns
fig, axes = plt.subplots(2, 3, figsize=(6, 4))

# Set spacing between plots
plt.subplots_adjust(hspace=0.5, wspace=0.5)

# Use for loop to create each subplot
for i in range(6):
    row = i // 3  # Calculate row index
    col = i % 3   # Calculate column index

     # Add text (2,3,i) centered at (0.5, 0.5)
    axes[row, col].text(0.5, 0.5, f'(2, 3, {i+1})', 
                       ha='center', va='center', 
                       fontsize=16, transform=axes[row, col].transAxes)

    # Set title for each subplot
    axes[row, col].set_title(f'Title {i+1}')

    # Set axis limits to match the image
    axes[row, col].set_xlim(0, 1)
    axes[row, col].set_ylim(0, 1)

plt.show()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(12,8))
fig.subplots_adjust(hspace=0.5, wspace=0.5)

for i, ax in enumerate(axes.flat, start=1):
    ax.set_title(f"Title {i}")
    ax.text(0.5, 0.5, f"(2,3,{i})", ha="center", va="center")

plt.suptitle('Subplots with Text')
plt.show()
```
