```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## create the DataFrame
df = pd.DataFrame({
    'name': ['christopher','marion','maria','mia','clement','randy','remi'],
    'age': [70,30,22,19,45,33,20],
    'gender': ['M','F','F','F','M','M','M'],
    'state': ['california','dc','california','dc','california','new york','porto'],
    'num_children': [2,0,0,3,8,1,4],
    'num_pets': [5,1,0,5,2,2,3]
})
```

```python
## create the bar plot
ax = df.plot(x='name', y='age', kind='bar', 
             title='Age per name', 
             legend=True, 
             figsize=(10, 6))

# Customize the plot
ax.set_xlabel('name')
ax.set_ylabel('age')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Number of children and pets per person
df.plot(x="name", y=["num_children","num_pets"], kind="bar", 
        figsize=(10, 6), 
        title="Number of Children & Pets per Person")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
# Create the another bar plot
ax = df.plot(x='name', y=['num_children', 'num_pets'], kind='bar', 
             title='Number of Children and Pets per Person', 
             figsize=(10, 6))
ax.set_xlabel('Name')
ax.legend(title='Category')
plt.tight_layout()
plt.show()
```

```python
# Create another bar plot
ax = df.plot(
    x='name',
    y=['num_children', 'num_pets'],
    kind='line',
    title='Number of Children and Pets per Person',
    figsize=(10, 6)
)
ax.set_xlabel('Name')
ax.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
"""
Key Learning Points:

    Pandas .plot() is a wrapper around matplotlib, making it easy to create quick visualizations
    kind parameter specifies the plot type ('bar', 'scatter', 'line', etc.)
    x and y parameters specify which columns to use for axes
    title, legend, figsize help customize the appearance
    Always consider data relationships when choosing visualization types
"""
```
