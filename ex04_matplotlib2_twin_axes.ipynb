```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data
left_data = [5, 7, 11, 13, 17]
right_data = [0.1, 0.2, 0.4, 0.8, -1.6]
x_axis = [0.0, 1.0, 2.0, 3.0, 4.0]

# Create figure and primary axis
fig, ax1 = plt.subplots(figsize=(8, 6))

# Plot left data (black line)
ax1.plot(x_axis, left_data, 'k-', linewidth=2)
ax1.set_xlabel('x_axis')
ax1.set_ylabel('Big', color='black')
ax1.tick_params(axis='y', labelcolor='black')

#Create secondary y-axis
ax2 = ax1.twinx()

# Plot right data (red line)
ax2.plot(x_axis, right_data, 'r-', linewidth=2)
ax2.set_ylabel('Small', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Set title
plt.title('Twin axes')

plt.tight_layout()
plt.show()
```

```python
"""
Matplotlib Key Concepts:
Plot Styling:

     Line styles: `-` (solid), `--` (dashed), `-.` (dashdot), `:` (dotted)
     Colors: `'r'` (red), `'b'` (blue), `'k'` (black), `'g'` (green)
     Markers: `'o'` (circle), `'s'` (square), `'^'` (triangle)
     Combine: `'r-.'` = red dashdot line

Advanced Features:

     `twinx()`: Create secondary y-axis
     `set_xlim()`, `set_ylim()`: Set axis limits
     `linewidth`, `markersize`: Control appearance
     `tick_params()`: Customize tick appearance

"""
```
