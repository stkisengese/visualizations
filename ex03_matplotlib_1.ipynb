```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data points (assumed integer coordinates)
x_data = [1, 4, 5, 6, 7]
y_data = [2, 6, 3, 6, 3]

# Create the plot
fig, ax = plt.subplots(figsize=(8, 6))

# Plot line with red dashdot style, width 3
ax.plot(x_data, y_data, 'r-.', linewidth=3, label='data')

# Plot blue circles with size 12
ax.plot(x_data, y_data, 'bo', markersize=12)

# Customize the plot
ax.set_title('Display marker')
ax.set_xlabel('x - axis')
ax.set_ylabel('y - axis')

# Set axis limits to [1,8]
ax.set_xlim(1, 8)
ax.set_ylim(1, 8)

# Add grid for better visualization
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Create the plot using matplotlib
plt.figure(figsize=(8, 6))

# Data points
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [1, 4, 7, 6, 3, 8, 2, 5]

# Plot with specified style
plt.plot(x, y, 'r-.', linewidth=3, label='Line')
plt.plot(x, y, 'bo', markersize=12, label='Points')

# Set plot properties
plt.title('Custom Matplotlib Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xlim(1, 8)
plt.ylim(1, 8)
plt.legend()
plt.grid(True, alpha=0.9)
plt.tight_layout()
plt.show()
```

```python
# Another plot - Quadratic Function
x = [1,2,3,4,5,6,7,8]
y = [1,4,9,16,25,36,49,64]

plt.plot(x, y, "r-.", linewidth=3, label="y=x")  # red dashdot
plt.scatter(x, y, color="blue", s=12**2, label="points")  # blue circles
plt.title("Quadratic Function")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.xlim([1,8])
plt.ylim([1,8**2])
plt.legend()
plt.show()
```

```python
# another custom plot

# Assume integer coordinates: let's use x = [1,2,...,8], y = some function
x = np.arange(1, 9)
y_line = x  # simple line
y_scatter = x[::-1]  # reverse for contrast

plt.figure(figsize=(8, 6))
plt.plot(x, y_line, 'r-.', linewidth=3, label='Red dashdot line')
plt.scatter(x, y_scatter, c='blue', s=12**2, label='Blue circles')  # s=size in points

plt.title('Custom Matplotlib Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xlim(1, 8)
plt.ylim(1, 8)
plt.legend()
plt.grid(True)
plt.show()
```
