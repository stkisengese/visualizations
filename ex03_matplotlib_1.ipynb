```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data points (assumed integer coordinates)
x_data = [1, 4, 5, 6, 7]
y_data = [2, 6, 3, 6, 3]

# Create the plot
fig, ax = plt.subplots(figsize=(8, 6))

# Plot line with red dashdot style, width 3
ax.plot(x_data, y_data, 'r-.', linewidth=3, label='data')

# Plot blue circles with size 12
ax.plot(x_data, y_data, 'bo', markersize=12)

# Customize the plot
ax.set_title('Display marker')
ax.set_xlabel('x - axis')
ax.set_ylabel('y - axis')

# Set axis limits to [1,8]
ax.set_xlim(1, 8)
ax.set_ylim(1, 8)

# Add grid for better visualization
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
